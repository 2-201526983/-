```python
def convert_day(x):
    if x == 0:
        return 'Mon'
    elif x == 1:
        return 'Tue'
    elif x == 2:
        return 'Wed'
    elif x == 3:
        return 'Thu'
    elif x == 4:
        return 'Fri'
    elif x == 5:
        return 'Sat'
    else:
        return 'Sun'
```

```python
df['요일'] = df['요일'].apply(convert_day)
```
